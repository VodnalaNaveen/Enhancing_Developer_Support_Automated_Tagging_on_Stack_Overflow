```python
import requests,re
from bs4 import BeautifulSoup;
import numpy as np
import pandas as pd
```

```python
x=['votes','frequent','trending','hotnotime']
que=[]
dis=[]
tagb = []
```

```python
for i in x:
    for j in range(1,10001):
        url=f'https://stackoverflow.com/questions?tab={i}&page={j}'
        response=requests.get(url,headers = {'User-Agent':'Mozilla/5.0'})
        page = BeautifulSoup(response.text)
        
        for q in page.find_all('h3',class_='s-post-summary--content-title'):
                if re.findall(r'">(\w.+?)<\/a>', str(q)):
                    que.append(''.join(re.findall(r'">(\w.+?)<\/a>', str(q))))
                else:
                     que.append(np.nan)
        
        for w in page.find_all('div',class_='s-post-summary--content-excerpt'):
                if re.findall(r'<div class="s-post-summary--content-excerpt">\s(.+\s+.+)', str(w)):
                    dis.append(''.join(re.findall(r'<div class="s-post-summary--content-excerpt">\s(.+\s+.+)', str(w))))

        html_string = str(page.find_all('ul',class_='ml0 list-ls-none js-post-tag-list-wrapper d-inline'))
        soup = BeautifulSoup(html_string, "html.parser")
        ul_blocks = soup.find_all("ul")

        
        for ul in ul_blocks:
            tags = [a.get_text(strip=True) for a in ul.find_all("a")]
            tagb.append(", ".join(tags)) 
```

```python
df = pd.DataFrame({'Questions':que,'Discription':dis,'Tags': tagb})
```
